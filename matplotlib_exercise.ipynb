```python
heart_disease=pd.read_csv("/content/heart-disease.csv")
```

```python
heart_disease.head(10)
```

```python
heart_disease['age'].plot.hist(figsize=(10,10),subplots=True,bins=50);
```

```python
heart_disease['age'].plot.hist(figsize=(10,30),subplots=True,bins=50);
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50
```

```python
fig,ax=plt.subplots(figsize=(10,6));
scatter=ax.scatter(over_50["age"],over_50['chol'],c=over_50['target']);

ax.set(title="Heart Disease and Cholesterol", xlabel="age",ylabel="cholesterol");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50['chol'].mean(),linestyle="--");
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-whitegrid")
fig,ax=plt.subplots(figsize=(10,6));
scatter=ax.scatter(over_50["age"],over_50['chol'],c=over_50['target']);

ax.set(title="Heart Disease and Cholesterol", xlabel="age",ylabel="cholesterol");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50['chol'].mean(),linestyle="--");
```

```python
fig,ax=plt.subplots(figsize=(10,6));
scatter=ax.scatter(over_50["age"],over_50['chol'],c=over_50['target'],cmap="winter");

ax.set(title="Heart Disease and Cholesterol", xlabel="age",ylabel="cholesterol");
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50['chol'].mean(),linestyle="--",c="red");
```
